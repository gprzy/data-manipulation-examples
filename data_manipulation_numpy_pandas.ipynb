# **PANDAS** (`pd`)

## **PARTE I**: Básico (`pd`)

### Criando tabelas com `pd.DataFrame()`

```python
df_cafes = pd.DataFrame(data={
    'id': [1, 2, 3, 4, 5, 6],
    'cafes': ['capuccino', 'mocha', 'machiato',
              'espresso', 'espresso lungo',
              'espresso stretto'],
    'intensidade': ['média', 'fraca', 'alta',
                    'alta', 'alta', 'alta']
})

df_cafes
```

```python
df_precos = pd.DataFrame(data=[[1, 22.50],
                               [2, 20.50],
                               [3, 23.50],
                               [4, 24.50],
                               [7, 30.50]], columns=['id', 'preco'])
df_precos
```

## **PARTE II**: Intermediário (`pd`)

### Mesclando tabelas com `pd.merge()`

```python
# destacar valores nulos na tabela
destacar_nulos = lambda x: 'background-color: yellow; color: black' \
                            if str(x) == 'nan' else ''
```

#### `INNER JOIN`

```python
# inner join
pd.merge(df_cafes, df_precos, on='id', how='inner')
```

#### `LEFT JOIN`

```python
# left join
left_join = pd.merge(df_cafes, df_precos, on='id', how='left')

# destacando a tabela
left_join = left_join.style.applymap(destacar_nulos)
left_join
```

#### `RIGHT JOIN`

```python
 # right join
right_join = pd.merge(df_cafes, df_precos, on='id', how='right')

# destacando a tabela
right_join = right_join.style.applymap(destacar_nulos)
right_join
```

#### `OUTER JOIN`

```python
# outer join
outer_join = pd.merge(df_cafes, df_precos, on='id', how='outer')

# destacando a tabela
outer_join = outer_join.style.applymap(destacar_nulos)
outer_join
```

### Mesclando tabelas com `pd.merge_asof()`

```python
# destacar a tabela com verde
destacar_para_verde = lambda x: 'background-color: #BAFFC9; color: black'

# destacar ids diferentes da tabela
def destacar_ids_diferentes(linha):
    id = linha.loc['id']
    if id > 4:
        cor = '#FFB3BA' # vermelho
    else:
        cor = '#BAFFC9' # verde
    return ['background-color: {}; color: black' \
            .format(cor) for coluna in linha]
```

```python
# inner join
merge = pd.merge(df_cafes, df_precos, on='id', how='inner')

# destacando a tabela
merge = merge.style.applymap(destacar_para_verde)
merge
```

```python
# merge asof
merge_asof = pd.merge_asof(df_cafes, df_precos)

# destacando a tabela
merge_asof = merge_asof.style.apply(destacar_ids_diferentes, axis=1)
merge_asof
```

```python
# não existem ids 5 e 6 na tabela de preços!
df_precos[df_precos['id'] > 4]
```

### Concatenando tabelas com `pd.concat()`

```python
# destacar colunas da tabela
def destacar_tabela_v(linha):
    elemento = str(linha.loc['preco'])
    if elemento != 'nan':
        cor = '#FFB3BA' # vermelho
    else:
        cor = '#BAFFC9' # verde
    return ['background-color: {}; color: black' \
            .format(cor) for coluna in linha]

# destacar linhas da tabela
def destacar_tabela_h(coluna):
    if coluna.name == 'preco' or coluna.name == 'id_precos':
        cor = '#FFB3BA' # vermelho
    else:
        cor = '#BAFFC9' # verde
    return ['background-color: {}; color: black' \
            .format(cor) for linha in coluna]
```

#### Concatenando pelo eixo horizontal (`axis=1`)

```python
# juntando a partir das linhas (axis=0)
df_concat = pd.concat([df_cafes, df_precos], axis=0)

# destacando a tabela
df_concat = df_concat.reset_index().style.apply(destacar_tabela_v, axis=1)
df_concat
```

#### Concatenando pelo eixo vertical (`axis=0`)

```python
# juntando a partir das colunas (axis=1)
df_concat = pd.concat([df_cafes, df_precos], axis=1)

# destacando a tabela
# renomeando as colunas para não repetir 'id'
df_concat.columns = ['id_cafes', 'cafes', 'intensidade', 'id_precos', 'preco']
df_concat = df_concat.reset_index().style.apply(destacar_tabela_h, axis=0)
df_concat
```

## Renomeando colunas com `df.rename()`

```python
# deixando as colunas em letra minúscula
df.rename(
    columns={coluna: coluna.lower() for coluna in df.columns}, inplace=True
)

# renomeando
df = df.rename(columns={
    'passengerid': 'id_passageiro',
    'survived': 'sobreviveu',
    'pclass': 'pclasse',
    'name': 'nome',
    'sex': 'sexo',
    'age': 'idade',
    'fare': 'tarifa',
    'cabin': 'cabine',
    'embarked': 'embarcou'
})
```

## Criando novas colunas

```python
# porcentagem da tarifa em relação ao total das tarifas
df['porcentagem_tarifa'] = df['tarifa'] / df['tarifa'].sum()
```

### `df.apply()`

```python
# exemplo de apply
# somando o id do passageiro com seu núm. de classe

df['soma_id_classe'] = df.apply(
    lambda linha: linha['id_passageiro'] + linha['pclasse'],
    axis=1
)
```

### `df.map()`

```python
# exemplo de map
# o passageiro sobreviveu?

df['sobreviveu_str'] = df['sobreviveu'].map(
    lambda elemento: 'sim' if elemento == 1 else 'não'
)
```

```python
# outro exemplo de map
# o passageiro é maior de idade?

df['maior_idade'] = df['idade'].map(
    lambda idade: 'sim' if idade >= 18 else 'não'
)
```

```python
# olhando novas colunas
# selecioando apenas as desejadas

df[['id_passageiro', 'pclasse', 'soma_id_classe',
    'idade', 'maior_idade', 'tarifa',
    'porcentagem_tarifa', 'sobreviveu', 'sobreviveu_str']].head()
```

## Realizando algumas consultas com `df.read_sql()` + condições e filtros

### Qual é o número total e a porcentagem de pessoas que sobreviveram?

```python
# total e porcentagem de pessoas que sobreviveram
print('sobreviveram\t',
      (df['sobreviveu'] == 1).sum())

print('porcentagem\t',
      (df['sobreviveu'] == 1).sum() / len(df['sobreviveu']))
```

```python
query = """SELECT COUNT(sobreviveu),
           CAST(COUNT(sobreviveu) AS FLOAT) /
               (SELECT COUNT(sobreviveu)
                FROM titanic) AS porcentagem
           FROM titanic
           WHERE sobreviveu = 1"""

pd.read_sql(query, conexao)
```

### Qual é a média de idade das mulheres que sobreviveram?

```python
# média de idade das mulheres que sobreviveram

condicao_mulher = (df['sexo'] == 'female')
condicao_sobreviveu = (df['sobreviveu'] == 1)

df[condicao_mulher & condicao_sobreviveu]['idade'].mean()
```

```python
query = """SELECT SUM(idade) / COUNT(idade) AS media_idade
           FROM titanic 
           WHERE sobreviveu = 1 AND sexo = 'female'"""

pd.read_sql(query, conexao)
```

### Homens ou mulheres morreram mais?

```python
# condições
condicao_homem = (df['sexo'] == 'male')
condicao_mulher = (df['sexo'] == 'female')
condicao_morreu = (df['sobreviveu'] == 0)

# df de homens que morreram e mulheres que morreram
df_homens_morreram = df[(condicao_homem & condicao_morreu)]
df_mulheres_morreram = df[(condicao_mulher & condicao_morreu)]

print(df_homens_morreram['id_passageiro'].count(),
      '\tnúmero de homens que morreram')

print(df_mulheres_morreram['id_passageiro'].count(),
      '\tnúmero de mulheres que morreram')
```

```python
query = """SELECT COUNT(id_passageiro) AS homens_morreram,
               (SELECT COUNT(id_passageiro) FROM titanic
                WHERE sobreviveu = 0 AND sexo = 'female') AS mulheres_morreram
           FROM titanic 
           WHERE sobreviveu = 0 AND sexo = 'male'"""

pd.read_sql(query, conexao)
```

### Qual foi a proporção entre homens e mulheres que morreram?

```python
# df de homens e homens que morreram
df_homens = df[condicao_homem]
df_homens_morreram = df[condicao_homem & condicao_morreu]

# df de mulheres e mulheres que morreram
df_mulheres = df[condicao_mulher]
df_mulheres_morreram = df[condicao_mulher & condicao_morreu]

print(df_homens_morreram['id_passageiro'].count() / len(df_homens),
      '\tporcentagem de homens que morreram')

print(df_mulheres_morreram['id_passageiro'].count() / len(df_mulheres),
      '\tporcentagem de mulheres que morreram')
```

```python
query = """SELECT CAST(COUNT(id_passageiro) AS FLOAT) /
                      (SELECT COUNT(id_passageiro) 
           FROM titanic WHERE sexo = 'male') AS porcentagem_homens,
               (SELECT COUNT(id_passageiro) / 
                       CAST((SELECT COUNT(id_passageiro) 
                             FROM titanic WHERE sexo = 'female') AS FLOAT) 
                FROM titanic
                WHERE sobreviveu = 0 AND sexo = 'female') AS porcentagem_mulheres
           FROM titanic 
           WHERE sobreviveu = 0 AND sexo = 'male'"""

pd.read_sql(query, conexao)
```
